### Title Extraction and Analysis Task

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
```

```python
# calculate survival rate
title_survival_rate = df.groupby('Title')['Survived'].mean()
print(title_survival_rate)
```

```python
#Mr: 0.156673, which means that 15.67% of the passengers with the "Mr." title survived.
#Mrs: 0.792000, which means that 79.20% of the passengers with the "Mrs." title survived.
#Countess: 1.000000, which means that all passengers with the "Countess" title survived.
```

#### Cabin Grouphy Task

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
df['CabinGroup'] = df['Cabin'].str[0]
# calculate the grouphy
cabin_group_stats = df.groupby('CabinGroup').agg(
    TotalPassengers=('PassengerId', 'count'),
    SurvivalRate=('Survived', 'mean')
)
print(cabin_group_stats)
```

### Family Size Impact Task

```python
# calculate the family size
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# categorized
df['FamilySizeCategory'] = pd.cut(df['FamilySize'], bins=[0, 1, 4, float('inf')], labels=['Single', 'Small', 'Large'])

# calculate the survival rate
family_size_survival_rate = df.groupby('FamilySizeCategory')['Survived'].mean()
print(family_size_survival_rate)
```

### Age Binning and Survival Analysis Task

```python
# creating age ranges
bins = [0, 10, 20, 30, 40, 50, 60, 70, float('inf')]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# calculate the survival rate
age_group_survival_rate = df.groupby('AgeGroup')['Survived'].mean()
print(age_group_survival_rate)
```

### Cross Tab task

```python
# create pivot table
crosstab = pd.pivot_table(df, values='Survived', index='Pclass', columns='Sex', aggfunc=['count', 'mean'])
print(crosstab)
```

**Female:**
- **Survivors:** 94
- **Survival Rate:** 96.81%

**Male:**
- **Survivors:** 122
- **Survival Rate:** 36.89%

***Comment:*** The survival rate for first-class female passengers is extremely high. This indicates that women and children were prioritized during the evacuation. In contrast, the survival rate for first-class male passengers is quite low.

### Impute Missing Age Task

```python
# impute missing age
median_age_per_title = df.groupby('Title')['Age'].median()
df['Age'] = df.apply(lambda x: median_age_per_title[x['Title']] if pd.isnull(x['Age']) else x['Age'], axis=1)
```

### Heat map Task

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Yalnızca sayısal sütunları seçiyoruz
numeric_df = df.select_dtypes(include=["int64", "float64"])

# Korelasyon matrisini oluştur
corr = numeric_df.corr()

# Görsel boyutu ayarlama
plt.figure(figsize=(10, 6))

# Heatmap çizimi
sns.heatmap(
    corr,                # korelasyon matrisi
    annot=True,          # hücrelerin içine değerleri yazdır
    fmt=".2f",           # ondalık format
    cmap="coolwarm",     # renk paleti
    linewidths=0.5,      # hücreler arası çizgi
    cbar=True            # renk skalasını göster
)

plt.title("Titanic Veri Seti - Korelasyon Isı Haritası", fontsize=14)
plt.show()
```

### Identify Duplicate Ticket Holders Task

```python
duplicate_tickets = df[df.duplicated(['Ticket'], keep=False)]
ticket_survival_rate = duplicate_tickets.groupby('Ticket')['Survived'].mean()
print(ticket_survival_rate)
```

Explanation of Correlation Values:
1.000000: Perfect positive correlation. All passengers with this ticket number survived.
0.666667: Strong positive correlation. A significant portion of passengers with this ticket survived.
0.000000: No correlation. This ticket number has no relationship with survival; there are no survivors among these passengers.
0.500000: Moderate positive correlation. Half of the passengers with this ticket survived.

### Analyze Fare per Person Task

```python
# calculate per person fare
df['FarePerPerson'] = df['Fare'] / df['FamilySize']

# calculate corra
fare_correlation = df[['FarePerPerson', 'Survived']].corr()
print(fare_correlation)
```

#### Feature Interaction: Class and Fare Task

```python
# Calculate fare quantile
df['FareQuantile'] = pd.qcut(df['Fare'], q=4, labels=['LowFare', 'MediumFare', 'HighFare', 'VeryHighFare'])

# Convert FareQuantile to string before concatenation
df['ClassFareCategory'] = df['Pclass'].astype(str) + '_' + df['FareQuantile'].astype(str)

# Calculate survival rate
class_fare_survival_rate = df.groupby('ClassFareCategory')['Survived'].mean()

# Print the survival rate
print(class_fare_survival_rate)
```

### Survival Rates by Class and Fare Category

- **1_HighFare:** 0.529412  
  Approximately 53% of first-class passengers who paid a high fare survived.

- **1_LowFare:** 0.000000  
  None of the first-class passengers who paid a low fare survived.

- **2_MediumFare:** 0.383721  
  About 38% of second-class passengers who paid a medium fare survived.
