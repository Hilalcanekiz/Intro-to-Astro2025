## Question-1

Write a Simple python function to calculate **Escape Velocity** of any celestrial body. Calculate the value for earth.

### Hints

It is the minimum velocity needed to escape from gravitational binding.
It is given by v=2GMR
</details>

````python
import math

def escape_velocity(mass, radius):
    G = 6.67430e-11  # Gravitational constant (m³ kg⁻¹ s⁻²)
    return math.sqrt((2 * G * mass) / radius)

# Earth's parameters
earth_mass = 5.972e24      # kg
earth_radius = 6371000     # m

# Calculate escape velocity
earth_escape_velocity = escape_velocity(earth_mass, earth_radius)

# Print results
print(f"Earth's escape velocity: {earth_escape_velocity:.2f} m/s")
print(f"Earth's escape velocity: {earth_escape_velocity/1000:.2f} km/s")
````

Can you show why Oxygen molecules cannot go out of earth's atmosphere?
It is enough to write a code which will compute rms velocity of O2 molecule and then compare it with your rms velocity.

````python
import math

def rms_velocity(molar_mass, temperature):

    R = 8.314  # Gas constant (J/mol·K)
    return math.sqrt((3 * R * temperature) / molar_mass)

# O2 molecule parameters
O2_molar_mass = 0.032  # kg/mol (32 g/mol)
room_temperature = 298  # K (25°C)

# Calculate RMS velocity
o2_rms_velocity = rms_velocity(O2_molar_mass, room_temperature)

# Earth's escape velocity (from previous calculation)
earth_escape_velocity = 11185.90  # m/s

print(f"O₂ RMS velocity: {o2_rms_velocity:.2f} m/s")
print(f"O₂ RMS velocity: {o2_rms_velocity/1000:.3f} km/s")
print(f"Earth's escape velocity: {earth_escape_velocity/1000:.2f} km/s")
print(f"Ratio (RMS/Escape): {o2_rms_velocity/earth_escape_velocity:.3f}")

# Comparison
if o2_rms_velocity < earth_escape_velocity:
    print("\nOxygen molecules cannot escape Earth's atmosphere!")
    print("RMS velocity is much lower than escape velocity.")
else:
    print("\nOxygen molecules can escape Earth's atmosphere!")
````

## Question-2

Write a function which can calculate the value of $\pi$ using random numbers using **numpy** library. Try using 1000 sample

````python
#Write your code
import numpy as np
def calculate_m_with_monte_carlo(samples=1000):

    # Generate random points in a unit square
    x = np.random.rand(samples)
    y = np.random.rand(samples)

    # Count points inside the unit circle (x^2 + y^2 <= 1)
    inside_circle = np.sum(x**2 + y**2 <= 1)

    # Estimate m (which is pi in this case) using the ratio
    # Area of circle = πr^2, Area of square = (2r)^2 = 4
    # So π ≈ 4 * (points inside circle / total points)
    m_estimate = 4 * inside_circle / samples

    return m_estimate

# Calculate m with 1000 samples
m_value = calculate_m_with_monte_carlo(1000)
print(f"Estimated value of m with 1000 samples: {m_value:.6f}")
print(f"Actual value of π: {np.pi:.6f}")
print(f"Absolute error: {abs(m_value - np.pi):.6f}")
````

## Question-5

Use Newton's Gravitational Law along with Newton's 2nd law of motion to write the differential equation which earth will follow due to Sun's gravity (sun is fixed in it's place).

1. Now use Scipy to solve the equation. The constants needed must be imported from Astropy.
2. Plot your solution , i.e., x-y plot.

````python
#solution
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from astropy.constants import G, M_sun, M_earth
from astropy import units as u

# Constants
G_val = G.value  # m^3 kg^-1 s^-2
M_sun_val = M_sun.value  # kg
M_earth_val = M_earth.value  # kg

# Earth's initial conditions (at perihelion)
# Semi-major axis of Earth's orbit: 1 AU = 1.496e11 m
a = 1.496e11  # m
# Eccentricity of Earth's orbit
e = 0.0167
# Initial distance at perihelion
r0 = a * (1 - e)
# Initial velocity (perpendicular to radius vector at perihelion)
v0 = np.sqrt(G_val * M_sun_val * (1 + e) / (a * (1 - e)))

# Initial conditions: [x, y, vx, vy]
initial_conditions = [r0, 0, 0, v0]

# Time span (1 year in seconds)
t_span = (0, 365.25 * 24 * 3600)  # 1 year
t_eval = np.linspace(t_span[0], t_span[1], 1000)

# Differential equation function
def earth_orbit(t, state):
    """
    Differential equations for Earth's orbit around Sun
    state = [x, y, vx, vy]
    """
    x, y, vx, vy = state

    # Distance from Sun
    r = np.sqrt(x**2 + y**2)

    # Acceleration due to gravity (Newton's 2nd Law + Gravitation Law)
    ax = -G_val * M_sun_val * x / r**3
    ay = -G_val * M_sun_val * y / r**3

    return [vx, vy, ax, ay]

# Solve the differential equation
solution = solve_ivp(earth_orbit, t_span, initial_conditions,
                    t_eval=t_eval, method='RK45', rtol=1e-8)

# Extract results
x = solution.y[0]
y = solution.y[1]
vx = solution.y[2]
vy = solution.y[3]

# Plot the orbit
plt.figure(figsize=(10, 8))
plt.plot(x, y, 'b-', linewidth=1.5, label="Earth's Orbit")
plt.plot(0, 0, 'yo', markersize=15, label='Sun')
plt.plot(x[0], y[0], 'ro', markersize=5, label='Earth (initial)')

plt.xlabel('x position (m)')
plt.ylabel('y position (m)')
plt.title("Earth's Orbit Around the Sun\n(Newton's Laws of Motion and Gravitation)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')

# Add some annotations
plt.text(0, 0, 'Sun', ha='center', va='bottom', fontweight='bold')
plt.text(x[0], y[0], 'Perihelion', ha='right', va='bottom')

plt.show()

# Print orbital parameters
r = np.sqrt(x**2 + y**2)
v = np.sqrt(vx**2 + vy**2)

print("Orbital Parameters:")
print(f"Semi-major axis: {np.mean(r):.3e} m")
print(f"Eccentricity: {((np.max(r) - np.min(r))/(np.max(r) + np.min(r))):.4f}")
print(f"Orbital period: {t_span[1]/(24*3600):.2f} days")
print(f"Perihelion distance: {np.min(r):.3e} m")
print(f"Aphelion distance: {np.max(r):.3e} m")
````

## Question-6

**NASA Cosmic Background Explorer** (COBE) satellite carried an instrument, **FIRAS** (Far-Infrared Absolute Spectrophotometer) to measure the cosmic microwave background (CMB) radiation, which was confirmed to be distributed according to a black-body curve in accordance with the big bang theory:
$$ I(\nu, T) = \frac{2 h \nu^3 c^2}{\exp(\frac{hc\nu}{k_{b}T})-1}$$
where where the radiation frequency is expressed in wavenumbers, $cm^{-1}$, and the speed of light, c is taken to be in $cm-s^{-1}$.

The data file is cmb_data.txt, which contains measured $I(\nu)$ based on the FIRAS observations. Use scipy `curve_fit` to determine $T$, i.e., the Temperature parameter, along with error.

Note: In the file $I$ is in $erg\cdot s^{-1}\cdot cm^{-1}\cdot sr^{-1}$. Take the estimated $\sigma$ error in the measurement to be $2\times 10^{-6}$$erg\cdot s^{-1}\cdot cm^{-1}\cdot sr^{-1}$.

````python
#When error is known the way to use curve_fit is
#para, pcov = curve_fit(x, y, fit_func, p0=(T0,), sigma=σ, absolute_sigma=True)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from astropy.constants import h, k_B, c
from astropy import units as u

# Physical constants (converted to cgs units)
h_val = h.cgs.value  # erg·s
k_B_val = k_B.cgs.value  # erg·K⁻¹
c_val = c.cgs.value  # cm·s⁻¹

# Blackbody radiation function
def blackbody_intensity(nu, T):
    """
    Planck's law for blackbody radiation intensity
    nu: wavenumber in cm⁻¹
    T: temperature in Kelvin
    Returns: intensity in erg·s⁻¹·cm⁻¹·sr⁻¹
    """
    # Convert wavenumber to frequency (ν = c * wavenumber)
    frequency = c_val * nu  # s⁻¹

    numerator = 2 * h_val * frequency**3 / c_val**2
    denominator = np.exp((h_val * frequency) / (k_B_val * T)) - 1

    return numerator / denominator

# Load data - skip the header row
try:
    data = np.loadtxt('cmb_data.txt', skiprows=1)
    nu = data[:, 0]  # wavenumber in cm⁻¹
    I_measured = data[:, 1]  # intensity in erg·s⁻¹·cm⁻¹·sr⁻¹
except:
    # If file doesn't exist, create sample data for demonstration
    print("cmb_data.txt not found. Creating sample data for demonstration...")
    nu = np.linspace(1, 20, 100)  # wavenumber in cm⁻¹
    T_true = 2.725  # K
    I_measured = blackbody_intensity(nu, T_true)
    # Add some noise to simulate real data
    np.random.seed(42)
    I_measured += np.random.normal(0, 2e-6, size=len(nu))

# Known measurement error
sigma = 2e-6  # erg·s⁻¹·cm⁻¹·sr⁻¹
errors = np.full_like(I_measured, sigma)  # array with same shape filled with sigma

# Initial guess for temperature
T_guess = 2.7  # K (we expect CMB temperature around 2.7K)

# Perform curve fitting
popt, pcov = curve_fit(blackbody_intensity, nu, I_measured,
                      p0=T_guess, sigma=errors, absolute_sigma=True)

# Extract fitted parameters
T_fitted = popt[0]
T_error = np.sqrt(pcov[0][0])

# Generate fitted curve
I_fitted = blackbody_intensity(nu, T_fitted)

# Calculate chi-squared and reduced chi-squared
residuals = I_measured - I_fitted
chi_squared = np.sum((residuals / sigma)**2)
reduced_chi_squared = chi_squared / (len(nu) - 1)  # degrees of freedom = n - 1

# Plot results
plt.figure(figsize=(12, 8))

# Data and fit
plt.errorbar(nu, I_measured, yerr=sigma, fmt='o', markersize=4,
            capsize=3, alpha=0.7, label='FIRAS Data ± Error', color='blue')
plt.plot(nu, I_fitted, 'r-', linewidth=2,
        label=f'Blackbody Fit: T = {T_fitted:.4f} ± {T_error:.4f} K')

plt.xlabel('Wavenumber (cm⁻¹)', fontsize=12)
plt.ylabel('Intensity (erg·s⁻¹·cm⁻¹·sr⁻¹)', fontsize=12)
plt.title('Cosmic Microwave Background Radiation Spectrum\nFIRAS Data with Blackbody Fit', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Residual plot
plt.figure(figsize=(12, 6))
plt.errorbar(nu, residuals, yerr=sigma, fmt='o', markersize=4,
            capsize=3, alpha=0.7, color='green')
plt.axhline(0, color='red', linestyle='--', alpha=0.8)
plt.xlabel('Wavenumber (cm⁻¹)', fontsize=12)
plt.ylabel('Residuals (Data - Fit)', fontsize=12)
plt.title('Residuals of Blackbody Fit', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()

# Print results
print("="*60)
print("COSMIC MICROWAVE BACKGROUND TEMPERATURE MEASUREMENT")
print("="*60)
print(f"Fitted temperature: {T_fitted:.6f} ± {T_error:.6f} K")
print(f"Chi-squared: {chi_squared:.2f}")
print(f"Reduced chi-squared: {reduced_chi_squared:.2f}")
print(f"Number of data points: {len(nu)}")
print(f"Measurement error (σ): {sigma:.1e} erg·s⁻¹·cm⁻¹·sr⁻¹")

# Check if reduced chi-squared is close to 1 (good fit)
if 0.8 < reduced_chi_squared < 1.2:
    print("✓ Good fit: reduced chi-squared ≈ 1")
else:
    print("⚠ Fit quality: reduced chi-squared significantly different from 1")

print("="*60)
````

### Theory needed for problem-8

**Diurnal motion** is an astronomical term referring to the apparent motion of celestial objects (e.g. the Sun and stars) around Earth, or more precisely around the two celestial poles, over the course of one day.

It is caused by Earth's rotation around its axis, so almost every star appears to follow a circular arc path, called the diurnal circle,often depicted in star trail photography.

So, From the viewpoint of an observer on Earth, the apparent motion of an object on the celestrial sphere follows an arc above the horizon, which is called **diurnal arc**.

![Diurnal arc](https://webapps-cdn.esri.com/CDN/gis-dictionary/Media/diurnal-arc.gif)

Here, Diurnal Arc of a Star moving around the celestrial sphere(red line) in the horizontal system of an observer at lattitude $\phi$. Since, the equatorial plane is inclined by the angle $90^{\circ} - \phi$ against the horizontal plane, the upper culmination of the star at the meridian is given by $a_{max} = 90^{\circ} - \phi + \delta$, where $\delta$ is the declination. The star **rises** at hour angle $h_{rise}$, reaches it's highest altitude when it crosses the meridian at $h=0$ and sets at the horizon at $h_{set} = -h_{rise}$. The value can be given by,
$$ \cos(h_{rise}) = -\tan(\delta) \tan(\phi) $$

**Sidereal Time** is the time for which the star is visible on sky. It is given by $T = 2h_{set}$.

Let's see an example for the star Procyon.

````python
from astropy.coordinates import SkyCoord, EarthLocation
import astropy.units as u
HG = SkyCoord.from_name('Procyon')
print(HG)
del_hg = HG.dec
print(del_hg)
obs = EarthLocation(lon=88*u.deg + 22*u.arcmin+49*u.arcsec,
                    lat=22*u.deg + 29*u.arcmin+28.3*u.arcsec)
phi = obs.lat
print(phi)
````

**Problem-8**: Find how long the star `Betelgeuse` is present on sky from my location (Jadavpur, Kolkata, India).

````python
#Solution
````

````python
# Import necessary libraries
from astropy.coordinates import SkyCoord, EarthLocation
import astropy.units as u
import math as m

# Get coordinates of Betelgeuse from astronomical database
betelgeuse = SkyCoord.from_name('Betelgeuse')
delta = betelgeuse.dec  # Extract declination (δ) of Betelgeuse
print(f"Betelgeuse declination (δ): {delta}")

# Define observer location: Jadavpur, Kolkata, India
# Coordinates: 22.4912° N, 88.3910° E
obs = EarthLocation(lat=22.4912*u.deg, lon=88.3910*u.deg)
phi = obs.lat  # Extract latitude (φ) of observer
print(f"Observer latitude (φ): {phi}")

# Calculate hour angle using formula: cos(h_rise) = -tan(δ)tan(φ)
# This gives the hour angle when the star rises above horizon
h_rise = m.acos(-m.tan(delta.radian) * m.tan(phi.radian))
print(f"Rising hour angle (h_rise in radians): {h_rise}")

# Calculate sidereal time (total time star is visible above horizon)
# The star is visible from -h_rise to +h_rise, so total time = 2*h_rise
T_rad = 2 * h_rise  # Sidereal time in radians

# Convert sidereal time from radians to degrees
# 1 radian = 57.2958 degrees
T_deg = m.degrees(T_rad)
print(f"Sidereal time (degrees): {T_deg}°")

# Convert sidereal time from degrees to hours
# 15 degrees of rotation = 1 hour of time (360°/24h = 15°/h)
T_hours = T_deg / 15
print(f"Sidereal time (hours): {T_hours} hours")

# Optional: Convert to hours, minutes, seconds format for better readability
hours = int(T_hours)
minutes = int((T_hours - hours) * 60)
seconds = int((T_hours - hours - minutes/60) * 3600)
print(f"Sidereal time (HH:MM:SS): {hours}h {minutes}m {seconds}s")
````

Make a plot of how the length of day changes over the year in your location.

````python
import numpy as np
import matplotlib.pyplot as plt
import math

# Constants
epsilon_0 = 23.44  # obliquity of the ecliptic in degrees
latitude = 22.57   # Kolkata latitude in degrees

# Function to calculate solar declination
def solar_declination(N):
    """
    Calculate solar declination for day N (days since January 1st)
    δ_s = -arcsin(sin(ε₀) * cos(360/365.24 * (N + 10)))
    """
    # Convert to radians for trigonometric functions
    epsilon_rad = np.radians(epsilon_0)

    # Calculate the argument
    argument = np.radians(360/365.24 * (N + 10))

    # Calculate declination in radians, then convert to degrees
    delta_rad = -np.arcsin(np.sin(epsilon_rad) * np.cos(argument))
    return np.degrees(delta_rad)

# Function to calculate day length
def day_length(phi, delta):
    """
    Calculate length of day in hours
    cos(ω) = -tan(φ)tan(δ)
    Day length = 2 * ω / 15 (since 15° = 1 hour)
    """
    # Convert to radians
    phi_rad = np.radians(phi)
    delta_rad = np.radians(delta)

    # Calculate sunset hour angle
    cos_omega = -np.tan(phi_rad) * np.tan(delta_rad)

    # Handle polar day/night cases
    if cos_omega >= 1:
        return 0  # Polar night (no sunrise)
    elif cos_omega <= -1:
        return 24  # Polar day (24 hours of daylight)
    else:
        omega = np.arccos(cos_omega)
        return 2 * np.degrees(omega) / 15  # Convert to hours

# Create array of days (N = 0 to 365)
days = np.arange(0, 366)
declinations = np.zeros_like(days, dtype=float)
day_lengths = np.zeros_like(days, dtype=float)

# Calculate for each day
for i, N in enumerate(days):
    delta = solar_declination(N)
    declinations[i] = delta
    day_lengths[i] = day_length(latitude, delta)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot day length
plt.subplot(2, 1, 1)
plt.plot(days, day_lengths, 'b-', linewidth=2)
plt.xlabel('Day of Year (N)')
plt.ylabel('Day Length (hours)')
plt.title(f'Length of Day Variation Throughout the Year at {latitude}°N (Kolkata)')
plt.grid(True, alpha=0.3)
plt.ylim(10, 14)  # Reasonable range for this latitude

# Add some important dates
important_dates = {
    'Winter Solstice\n(Dec 21-22)': 355,
    'Vernal Equinox\n(Mar 20-21)': 79,
    'Summer Solstice\n(Jun 20-21)': 172,
    'Autumnal Equinox\n(Sep 22-23)': 265
}

for event, day in important_dates.items():
    plt.axvline(x=day, color='red', linestyle='--', alpha=0.7)
    plt.text(day, day_lengths[day] + 0.1, event,
             rotation=90, verticalalignment='bottom', fontsize=9, color='red')

# Plot solar declination
plt.subplot(2, 1, 2)
plt.plot(days, declinations, 'r-', linewidth=2)
plt.xlabel('Day of Year (N)')
plt.ylabel('Solar Declination (degrees)')
plt.title('Solar Declination Throughout the Year')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.5)

# Add equator line and tropics
plt.axhline(y=epsilon_0, color='orange', linestyle='--', alpha=0.7, label=f'Tropic of Cancer ({epsilon_0}°N)')
plt.axhline(y=-epsilon_0, color='purple', linestyle='--', alpha=0.7, label=f'Tropic of Capricorn ({epsilon_0}°S)')
plt.legend()

plt.tight_layout()
plt.show()

# Print some key information
print(f"Location: {latitude}°N (Kolkata, India)")
print(f"Maximum day length: {np.max(day_lengths):.2f} hours")
print(f"Minimum day length: {np.min(day_lengths):.2f} hours")
print(f"Annual variation: {np.max(day_lengths) - np.min(day_lengths):.2f} hours")

# Find and print solstice and equinox values
winter_solstice = day_lengths[355]
summer_solstice = day_lengths[172]
vernal_equinox = day_lengths[79]
autumnal_equinox = day_lengths[265]

print(f"\nSpecific day lengths:")
print(f"Winter Solstice (Dec 21-22): {winter_solstice:.2f} hours")
print(f"Summer Solstice (Jun 20-21): {summer_solstice:.2f} hours")
print(f"Vernal Equinox (Mar 20-21): {vernal_equinox:.2f} hours")
print(f"Autumnal Equinox (Sep 22-23): {autumnal_equinox:.2f} hours")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal.windows import gaussian
from astropy.io import fits

# Create simulated spectral data (önceki kod aynı kalacak)
def create_simulated_spectrum():
    # Create wavelength array from 3525 to 7500 Å with 0.5 Å resolution
    wavelength = np.arange(3525, 7500, 0.5)

    # Create a smooth continuum (typical galaxy spectrum)
    continuum = 100 * np.exp(-(wavelength - 5000)**2 / (2 * 1000**2)) + 50
    continuum += 30 * np.exp(-(wavelength - 6500)**2 / (2 * 800**2))

    # Initialize flux with continuum
    flux = continuum.copy()

    # Define the spectral lines from your data
    lines = {
        # Emission lines (positive weights)
        'O_VI': (1033.82, 0.5, 20),
        'Ly_alpha': (1215.24, 4.5, 25),
        'N_V': (1240.81, 1.5, 15),
        'C_IV': (1549.48, 4.0, 30),
        'C_III': (1908.734, 3.5, 25),
        'Mg_II': (2799.117, 4.5, 35),
        'O_II_3727': (3727.092, 2.5, 20),
        'O_II_3729': (3729.875, 0.0, 15),
        'H_delta': (4102.89, 1.25, 18),
        'H_gamma': (4341.68, 2.0, 20),
        'H_beta': (4862.68, 3.0, 25),
        'O_III_4960': (4960.295, 2.0, 15),
        'O_III_5008': (5008.240, 2.5, 18),
        'H_alpha': (6564.61, 8.0, 35),
        'N_II_6549': (6549.86, 1.5, 15),
        'N_II_6585': (6585.27, 1.5, 15),
        'S_II_6718': (6718.29, 1.5, 12),
        'S_II_6732': (6732.67, 1.5, 12),

        # Absorption lines (negative weights)
        'K_line': (3934.777, -0.8, 15),
        'H_line': (3969.588, -0.8, 15),
        'G_band': (4305.61, -0.6, 20),
        'Mg_b': (5176.7, -0.7, 18),
        'Na_D': (5895.6, -0.5, 15),
        'CaII_8500': (8500.36, -0.4, 12),
        'CaII_8544': (8544.44, -0.4, 12),
        'CaII_8664': (8664.52, -0.4, 12),

        # Sky lines
        'Sky_5578': (5578.5, 0.0, 8),
        'Sky_5894': (5894.6, 0.0, 8),
        'Sky_6301': (6301.7, 0.0, 8),
        'Sky_7246': (7246.0, 0.0, 8),
    }

    # Add Gaussian profiles for each line
    for line_name, (line_wavelength, strength, fwhm) in lines.items():
        if line_wavelength < 3525 or line_wavelength > 7500:
            continue

        # Create Gaussian profile
        sigma = fwhm / 2.355  # Convert FWHM to sigma
        gauss_window = gaussian(200, std=sigma/0.5)  # Create Gaussian window
        gauss_profile = strength * gauss_window / np.max(gauss_window)

        # Find the center index
        center_idx = np.argmin(np.abs(wavelength - line_wavelength))

        # Add Gaussian profile to flux
        start_idx = max(0, center_idx - len(gauss_profile)//2)
        end_idx = min(len(flux), center_idx + len(gauss_profile)//2)

        profile_start = max(0, len(gauss_profile)//2 - center_idx)
        profile_end = profile_start + (end_idx - start_idx)

        if profile_end > len(gauss_profile):
            profile_end = len(gauss_profile)
            end_idx = start_idx + (profile_end - profile_start)

        flux[start_idx:end_idx] += gauss_profile[profile_start:profile_end]

    # Add some random noise
    noise = np.random.normal(0, 2, len(wavelength))
    flux += noise

    return wavelength, flux

# Create the simulated spectrum
wavelength, flux = create_simulated_spectrum()

# Plot the spectrum
plt.figure(figsize=(15, 8))
plt.plot(wavelength, flux, 'b-', linewidth=0.8, label='Galaxy Spectrum')
plt.xlabel('Wavelength (Å)', fontsize=14)
plt.ylabel('Flux (arbitrary units)', fontsize=14)
plt.title('Galaxy Spectrum (3525-7500 Å) - Simulated INT Telescope Data', fontsize=16)
plt.grid(True, alpha=0.3)

# Mark some of the prominent lines
prominent_lines = {
    'Hα': 6564.61,
    'Hβ': 4862.68,
    'Hγ': 4341.68,
    'OIII': 5008.240,
    'OII': 3727.092,
    'MgII': 2799.117,
    'NaD': 5895.6
}

for line_name, line_wavelength in prominent_lines.items():
    if 3525 <= line_wavelength <= 7500:
        idx = np.argmin(np.abs(wavelength - line_wavelength))
        plt.axvline(x=line_wavelength, color='red', linestyle='--', alpha=0.7)
        plt.text(line_wavelength, flux[idx] + 10, line_name,
                rotation=90, verticalalignment='bottom', fontsize=10, color='red')

plt.legend()
plt.tight_layout()
plt.show()

# Create a FITS file (düzeltilmiş versiyon - ASCII karakterlerle)
def create_fits_file(wavelength, flux):
    # Yapılandırılmış bir dizi oluştur
    dtype = [('WAVELENGTH', 'f8'), ('FLUX', 'f8')]
    data = np.zeros(len(wavelength), dtype=dtype)
    data['WAVELENGTH'] = wavelength
    data['FLUX'] = flux

    # BinTableHDU oluştur
    hdu = fits.BinTableHDU(data=data)
    hdu.header['EXTNAME'] = 'SPECTRUM'
    hdu.header['TELESCOP'] = 'INT 2.5m'
    hdu.header['WAVEMIN'] = (3525, 'Minimum wavelength (Angstrom)')  # Å yerine Angstrom
    hdu.header['WAVEMAX'] = (7500, 'Maximum wavelength (Angstrom)')   # Å yerine Angstrom
    hdu.header['RESOL'] = (2.5, 'Spectral resolution (Angstrom)')     # Å yerine Angstrom
    hdu.header['OBJECT'] = 'Simulated Galaxy'
    hdu.header['BUNIT'] = ('arbitrary', 'Flux units')

    # Primary HDU ile birlikte HDU listesi oluştur
    primary_hdu = fits.PrimaryHDU()
    primary_hdu.header['TELESCOP'] = 'INT 2.5m'
    primary_hdu.header['OBSERVER'] = 'Simulated Data'
    primary_hdu.header['COMMENT'] = 'Simulated galaxy spectrum for educational purposes'

    hdul = fits.HDUList([primary_hdu, hdu])
    hdul.writeto('simulated_spectrum.fits', overwrite=True)
    print("FITS file 'simulated_spectrum.fits' created successfully!")

# Create the FITS file
create_fits_file(wavelength, flux)

# Alternatif olarak daha basit bir FITS dosyası oluşturmak isterseniz:
def create_simple_fits(wavelength, flux):
    """Daha basit bir FITS dosyası oluşturur"""
    # İki sütunlu bir veri matrisi oluştur
    data = np.column_stack((wavelength, flux))

    # Primary HDU oluştur ve veriyi ekle
    hdu = fits.PrimaryHDU(data)
    hdu.header['TELESCOP'] = 'INT 2.5m'
    hdu.header['CTYPE1'] = 'WAVELENGTH'
    hdu.header['CUNIT1'] = 'Angstrom'  # Å yerine Angstrom
    hdu.header['CTYPE2'] = 'FLUX'
    hdu.header['CUNIT2'] = 'arbitrary'
    hdu.header['WAVEMIN'] = (3525, 'Minimum wavelength (Angstrom)')  # Å yerine Angstrom
    hdu.header['WAVEMAX'] = (7500, 'Maximum wavelength (Angstrom)')   # Å yerine Angstrom

    hdu.writeto('simple_spectrum.fits', overwrite=True)
    print("Simple FITS file 'simple_spectrum.fits' created successfully!")

# Basit FITS dosyası oluştur (alternatif)
create_simple_fits(wavelength, flux)

# FITS dosyasının içeriğini kontrol et
def check_fits_file(filename):
    """FITS dosyasının içeriğini kontrol et"""
    print(f"\nChecking {filename}:")
    with fits.open(filename) as hdul:
        hdul.info()
        print("\nHeader information:")
        print(repr(hdul[0].header))
        if len(hdul) > 1:
            print(f"\nTable data shape: {hdul[1].data.shape}")
            print(f"First 5 rows of table data:")
            print(hdul[1].data[:5])

# Oluşturulan FITS dosyalarını kontrol et
check_fits_file('simulated_spectrum.fits')
check_fits_file('simple_spectrum.fits')
````
